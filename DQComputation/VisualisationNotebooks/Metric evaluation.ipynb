# missing

```python
iris_m_qa, iris_m_qf, _ = np.load("../Output/scores/iris_missing_(qa,qf,time).npy", allow_pickle=True)
cancer_m_qa, cancer_m_qf, _ = np.load("../Output/scores/cancer_missing_(qa,qf,time).npy", allow_pickle=True)
adult_m_qa, adult_m_qf, _ = np.load("../Output/scores/adult_missing_(qa,qf,time).npy", allow_pickle=True)

abalone_m_qa, abalone_m_qf, _ = np.load("../Output/scores/abalone_missing_(qa,qf,time).npy", allow_pickle=True)
statlog_m_qa, statlog_m_qf, _ = np.load("../Output/scores/statlog_missing_(qa,qf,time).npy", allow_pickle=True)
spambase_m_qa, spambase_m_qf, _ = np.load("../Output/scores/spambase_missing_(qa,qf,time).npy", allow_pickle=True)
heart_m_qa, heart_m_qf, _ = np.load("../Output/scores/heart_missing_(qa,qf,time).npy", allow_pickle=True)
bean_m_qa, bean_m_qf, _ = np.load("../Output/scores/bean_missing_(qa,qf,time).npy", allow_pickle=True)
```

```python
ys = [0, 0.3, 0.6]
heights = [0.3, 0.3, 0.4]
colors = cm.rainbow([0.5, 0.75, 1])
zones = []

for j in range(3):
    zones.append(patches.Rectangle((0, ys[j]), 55, heights[j], color=colors[j], alpha=0.33))

fig, ax = plt.subplots()
x = np.array(range(0, 55, 5))

line1a, = plt.plot(x, iris_m_qa[0], color='C0')
line2a, = plt.plot(x, cancer_m_qa[0], color='C1')
line3a, = plt.plot(x, adult_m_qa[0], color='C2')

line1f, = plt.plot(x, iris_m_qf[0], '--', color='C0')
line2f, = plt.plot(x, cancer_m_qf[0], '--', color='C1')
line3f, = plt.plot(x, adult_m_qf[0], '--', color='C2')

plt.ylim([0, 1])
plt.xlim([0, 50])

ymajor_ticks = np.arange(0, 11, 1)/10
yminor_ticks = np.arange(0, 20, 1)/20
ax.set_yticks(ymajor_ticks)
ax.set_yticks(yminor_ticks, minor=True)
xmajor_ticks = np.arange(0, 51, 10)
xminor_ticks = np.arange(0, 50, 5)
ax.set_xticks(xmajor_ticks)
ax.set_xticks(xminor_ticks, minor=True)
ax.grid(which='minor', linestyle='--')
legend1 = plt.legend([line1a, line2a, line3a], ['Iris', 'Breast Cancer', 'Adult'])
plt.gca().add_artist(legend1)

pc = PatchCollection(zones, match_original=True)
ax.add_collection(pc)
plt.grid(alpha=0.5)
fig.tight_layout()
plt.show()
```
